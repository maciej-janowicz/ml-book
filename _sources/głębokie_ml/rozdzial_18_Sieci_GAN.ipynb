```python
# -*- coding: utf-8 -*-
"""
# Sieci GAN (Generative Adversarial Networks) w Bran≈ºy Ubezpieczeniowej
## Kompleksowy przewodnik z praktycznymi zastosowaniami

### Autor: Prezentacja dla firmy ubezpieczeniowej
### Data: 2025

---

## 1. WPROWADZENIE DO SIECI GAN

### Co to sƒÖ sieci GAN?

Generative Adversarial Networks (GAN) to architektura sztucznej inteligencji,
kt√≥ra sk≈Çada siƒô z dw√≥ch sieci neuronowych "walczƒÖcych" ze sobƒÖ:

1. **Generator** - tworzy sztuczne dane pr√≥bujƒÖc oszukaƒá dyskryminator
2. **Dyskryminator** - stara siƒô odr√≥≈ºniƒá prawdziwe dane od sztucznych

Ta "gra" miƒôdzy sieciami prowadzi do powstania generatora, kt√≥ry potrafi tworzyƒá
bardzo realistyczne sztuczne dane.

### Dlaczego GAN sƒÖ u≈ºyteczne dla ubezpieczycieli?

- **Augmentacja danych**: Tworzenie dodatkowych danych treningowych
- **Symulacja scenariuszy**: Modelowanie rzadkich zdarze≈Ñ ubezpieczeniowych
- **Ochrona prywatno≈õci**: Generowanie syntetycznych danych klient√≥w
- **Wykrywanie oszustw**: Tworzenie przyk≈Çad√≥w fraud√≥w do treningu modeli
- **Optymalizacja produkt√≥w**: Symulacja zachowa≈Ñ r√≥≈ºnych segment√≥w klient√≥w

Nieco dok≈Çadniej:

1. Augmentacja danych (szczeg√≥lnie rzadkich przypadk√≥w oszustw)
   - W danych ubezpieczeniowych fraud to zjawisko rzadkie (np. 1‚Äì3% przypadk√≥w).
   - To powoduje, ≈ºe klasyfikatory uczƒÖ siƒô na bardzo niezr√≥wnowa≈ºonych zbiorach.
   - GAN-y mogƒÖ generowaƒá sztuczne pr√≥bki transakcji/fraud√≥w, kt√≥re wyglƒÖdajƒÖ realistycznie
     i pozwalajƒÖ trenowaƒá modele wykrywania oszustw.
   - Przyk≈Çad: wygenerowanie dodatkowych 1000 ‚Äûfa≈Çszywych wniosk√≥w o rentƒô‚Äù
     o cechach podobnych do znanych fraud√≥w.

2. Symulacja nietypowych scenariuszy
   - GAN potrafi tworzyƒá ‚Äûsyntetyczne‚Äù dane odpowiadajƒÖce rzadkim, ekstremalnym zdarzeniom.
   - W ubezpieczeniach mo≈ºe to byƒá np.:
     - nietypowy schemat wy≈Çudze≈Ñ w ≈õwiadczeniach d≈Çugoterminowych,
     - fa≈Çszywe faktury medyczne z drobnymi wariacjami,
     - nietypowe profile klient√≥w w analizie ryzyka.
   To pozwala trenowaƒá systemy na oszustwach, kt√≥re jeszcze siƒô nie wydarzy≈Çy, ale sƒÖ prawdopodobne.

3. Detekcja anomalii poprzez ‚Äûdyskryminator‚Äù
   - W klasycznym GAN mamy generator i dyskryminator.
   - Dyskryminator uczy siƒô odr√≥≈ºniaƒá ‚Äûrealne dane‚Äù od ‚Äûpodrobionych‚Äù.
   - W zastosowaniu do oszustw mo≈ºna wykorzystaƒá sam dyskryminator jako detektor anomalii:
   - Je≈õli nowy wniosek wyglƒÖda ‚Äûnienaturalnie‚Äù wzglƒôdem zbioru uczciwych wniosk√≥w, zostaje oznaczony jako podejrzany.

4. Weryfikacja dokument√≥w (tekst/obraz)
   - GAN-y dobrze radzƒÖ sobie z danymi obrazowymi, wiƒôc mogƒÖ byƒá u≈ºywane np. do:
     - wykrywania fa≈Çszywych skan√≥w dokument√≥w (dowody osobiste, za≈õwiadczenia lekarskie, akty urodzenia/zgonu),
     - rozpoznawania subtelnych manipulacji (deepfake w dokumentach).

5. Ochrona prywatno≈õci (synthetic data)
   - W ubezpieczeniach dane osobowe sƒÖ bardzo wra≈ºliwe (RODO).
   - GAN-y mogƒÖ generowaƒá syntetyczne bazy danych klient√≥w,
   - kt√≥re statystycznie odwzorowujƒÖ prawdziwe dane, ale nie zawierajƒÖ danych realnych os√≥b.
   - Takie bazy mo≈ºna bezpiecznie udostƒôpniaƒá do test√≥w i prac rozwojowych.

6. Czynniki potencjalneego ryzyka
   - GAN-y mogƒÖ byƒá te≈º u≈ºyte przez oszust√≥w do generowania bardzo przekonujƒÖcych
     fa≈Çszywych dokument√≥w czy deepfake‚Äô√≥w - instytucje muszƒÖ byƒá przygotowane na obronƒô przed tym.

Uwaga: w swoim najwa≈ºniejszym obszarze zastosowa≈Ñ, mianowicie w grafice,
GAN-y spotka≈Çy siƒô w ostatnich czterech latach z silnƒÖ konkurencjƒÖ modeli dyfuzyjnych.
PosiadajƒÖ one jednak istotne zalety, dziƒôki kt√≥rym nie zosta≈Çy wyparte z tych
zastosowa≈Ñ; najwa≈ºniejsze z nich to szybko≈õƒá i zdolno≈õƒá do skutecznej pracy
na ma≈Çych zbiorach danych.
"""

# Instalacja wymaganych bibliotek (uruchom w pierwszej kolejno≈õci)
!pip install torch torchvision matplotlib numpy pandas scikit-learn seaborn

import torch
import torch.nn as nn
import torch.optim as optim
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import make_classification
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

# Ustawienie seedu dla reprodukowalno≈õci
torch.manual_seed(42)
np.random.seed(42)

print("‚úÖ Biblioteki za≈Çadowane pomy≈õlnie!")
print("üìä Przygotowano ≈õrodowisko do nauki o sieciach GAN")

"""
---

## 2. PRZYK≈ÅAD 1: PODSTAWOWA STRUKTURA GAN
### Poziom trudno≈õci: ‚≠ê (Podstawowy)

Zaczniemy od prostego przyk≈Çadu pokazujƒÖcego podstawowƒÖ strukturƒô sieci GAN.
Stworzymy generator i dyskryminator dla jednowymiarowych danych.
"""

class SimpleGenerator(nn.Module):
    """
    Prosty generator do tworzenia jednowymiarowych danych.

    W kontek≈õcie ubezpieczeniowym mo≈ºe to byƒá np. generator
    wysoko≈õci szk√≥d lub wieku klient√≥w.
    """
    def __init__(self, input_size=10, output_size=1):
        super(SimpleGenerator, self).__init__()

        # Sieƒá sk≈ÇadajƒÖca siƒô z trzech warstw liniowych
        # input_size: rozmiar wektora szumu (noise)
        # output_size: rozmiar generowanych danych
        self.network = nn.Sequential(
            nn.Linear(input_size, 128),    # Warstwa wej≈õciowa
            nn.ReLU(),                     # Funkcja aktywacji ReLU
            nn.Linear(128, 64),            # Warstwa ukryta
            nn.ReLU(),
            nn.Linear(64, output_size),    # Warstwa wyj≈õciowa
            nn.Tanh()                      # Tanh daje warto≈õci [-1, 1]
        )

    def forward(self, x):
        """Przepuszczenie danych przez sieƒá"""
        return self.network(x)

class SimpleDiscriminator(nn.Module):
    """
    Prosty dyskryminator oceniajƒÖcy czy dane sƒÖ prawdziwe czy sztuczne.

    W ubezpieczeniach mo≈ºe oceniaƒá czy profil klienta jest autentyczny.
    """
    def __init__(self, input_size=1):
        super(SimpleDiscriminator, self).__init__()

        # Sieƒá klasyfikujƒÖca dane jako prawdziwe (1) lub fa≈Çszywe (0)
        self.network = nn.Sequential(
            nn.Linear(input_size, 64),     # Warstwa wej≈õciowa
            nn.LeakyReLU(0.2),            # LeakyReLU lepsze dla discriminatora
            nn.Linear(64, 32),             # Warstwa ukryta
            nn.LeakyReLU(0.2),
            nn.Linear(32, 1),              # Warstwa wyj≈õciowa (1 neuron)
            nn.Sigmoid()                   # Sigmoid daje prawdopodobie≈Ñstwo [0,1]
        )

    def forward(self, x):
        """Przepuszczenie danych przez sieƒá"""
        return self.network(x)

# Tworzenie instancji sieci
generator = SimpleGenerator()
discriminator = SimpleDiscriminator()

print("üèóÔ∏è  Utworzono podstawowe sieci GAN:")
print(f"Generator: {sum(p.numel() for p in generator.parameters())} parametr√≥w")
print(f"Discriminator: {sum(p.numel() for p in discriminator.parameters())} parametr√≥w")

# Testowanie sieci
noise = torch.randn(5, 10)  # 5 pr√≥bek szumu, ka≈ºda 10-wymiarowa
fake_data = generator(noise)
prediction = discriminator(fake_data)

print(f"\nüìä Test dzia≈Çania:")
print(f"Wygenerowane dane: {fake_data.squeeze().detach().numpy()}")
print(f"Ocena discriminatora: {prediction.squeeze().detach().numpy()}")

"""
### üí° Zastosowanie w ubezpieczeniach:

Ten podstawowy model mo≈ºe byƒá rozwiniƒôty do:
- Generowania syntetycznych warto≈õci wieku klient√≥w do testowania modeli
- Tworzenia rozk≈Çad√≥w szk√≥d dla r√≥≈ºnych produkt√≥w ubezpieczeniowych
- Symulacji podstawowych parametr√≥w ryzyka

---

## 3. PRZYK≈ÅAD 2: GAN DO GENEROWANIA DANYCH KLIENT√ìW UBEZPIECZENIOWYCH
### Poziom trudno≈õci: ‚≠ê‚≠ê (≈Åatwy)

Teraz stworzymy bardziej praktyczny przyk≈Çad - GAN generujƒÖcy profile klient√≥w ubezpieczeniowych.
"""

def create_insurance_dataset(n_samples=1000):
    """
    Tworzy syntetyczny zbi√≥r danych klient√≥w ubezpieczeniowych.

    Cechy klient√≥w:
    - Wiek (25-80 lat)
    - Doch√≥d roczny (20000-200000)
    - Liczba wypadk√≥w w historii (0-5)
    - Sta≈º jazdy w latach (5-50)
    """
    np.random.seed(42)

    # Generowanie realistycznych danych klient√≥w
    age = np.random.normal(45, 15, n_samples)
    age = np.clip(age, 25, 80)  # Ograniczenie wieku do rozsƒÖdnych warto≈õci

    # Doch√≥d koreluje z wiekiem (starsi = wy≈ºsze zarobki)
    income = 30000 + age * 1000 + np.random.normal(0, 20000, n_samples)
    income = np.clip(income, 20000, 200000)

    # Liczba wypadk√≥w - rzadkie zdarzenia
    accidents = np.random.poisson(0.5, n_samples)
    accidents = np.clip(accidents, 0, 5)

    # Sta≈º jazdy zwiƒÖzany z wiekiem
    driving_experience = age - 18 + np.random.normal(0, 3, n_samples)
    driving_experience = np.clip(driving_experience, 5, 100)

    # ≈ÅƒÖczenie danych
    data = np.column_stack([age, income, accidents, driving_experience])

    return data

# Tworzenie zbioru danych
real_data = create_insurance_dataset(1000)
feature_names = ['Wiek', 'Doch√≥d', 'Wypadki', 'Sta≈º jazdy']

# Normalizacja danych (wa≈ºne dla GAN!)
scaler = StandardScaler()
real_data_normalized = scaler.fit_transform(real_data)

print("üìã Przyk≈Çadowe dane klient√≥w ubezpieczeniowych:")
df = pd.DataFrame(real_data, columns=feature_names)
print(df.head(10))

# Wizualizacja rozk≈Çad√≥w
fig, axes = plt.subplots(2, 2, figsize=(12, 8))
fig.suptitle('Rozk≈Çady cech klient√≥w ubezpieczeniowych', fontsize=16)

for i, (feature, ax) in enumerate(zip(feature_names, axes.flat)):
    ax.hist(real_data[:, i], bins=30, alpha=0.7, color='skyblue')
    ax.set_title(f'Rozk≈Çad: {feature}')
    ax.set_xlabel(feature)
    ax.set_ylabel('Czƒôsto≈õƒá')

plt.tight_layout()
plt.show()

class InsuranceGenerator(nn.Module):
    """
    Generator profili klient√≥w ubezpieczeniowych.

    Generuje 4-wymiarowe wektory reprezentujƒÖce:
    wiek, doch√≥d, liczbƒô wypadk√≥w, sta≈º jazdy
    """
    def __init__(self, noise_dim=100, output_dim=4):
        super(InsuranceGenerator, self).__init__()

        # Wiƒôksza sieƒá dla bardziej z≈Ço≈ºonych danych
        self.network = nn.Sequential(
            nn.Linear(noise_dim, 128),
            nn.BatchNorm1d(128),          # Batch normalization stabilizuje trening
            nn.ReLU(),

            nn.Linear(128, 256),
            nn.BatchNorm1d(256),
            nn.ReLU(),

            nn.Linear(256, 128),
            nn.BatchNorm1d(128),
            nn.ReLU(),

            nn.Linear(128, output_dim),
            nn.Tanh()                     # Dane znormalizowane do [-1, 1]
        )

    def forward(self, noise):
        return self.network(noise)

class InsuranceDiscriminator(nn.Module):
    """
    Dyskryminator oceniajƒÖcy autentyczno≈õƒá profili klient√≥w.
    """
    def __init__(self, input_dim=4):
        super(InsuranceDiscriminator, self).__init__()

        self.network = nn.Sequential(
            nn.Linear(input_dim, 128),
            nn.LeakyReLU(0.2),
            nn.Dropout(0.3),              # Dropout przeciwdzia≈Ça overfitting

            nn.Linear(128, 64),
            nn.LeakyReLU(0.2),
            nn.Dropout(0.3),

            nn.Linear(64, 32),
            nn.LeakyReLU(0.2),

            nn.Linear(32, 1),
            nn.Sigmoid()
        )

    def forward(self, data):
        return self.network(data)

# Inicjalizacja sieci dla danych ubezpieczeniowych
gen_insurance = InsuranceGenerator()
disc_insurance = InsuranceDiscriminator()

print(f"\nüèóÔ∏è  Utworzono GAN dla danych ubezpieczeniowych:")
print(f"Generator: {sum(p.numel() for p in gen_insurance.parameters())} parametr√≥w")
print(f"Dyskryminator: {sum(p.numel() for p in disc_insurance.parameters())} parametr√≥w")

# Test generowania
test_noise = torch.randn(5, 100)
test_profiles = gen_insurance(test_noise)
test_predictions = disc_insurance(test_profiles)

print(f"\nüß™ Test generowania profili klient√≥w:")
print("Wygenerowane profile (znormalizowane):")
print(test_profiles.detach().numpy())

"""
### üíº Korzy≈õci biznesowe dla ubezpieczyciela:

1. **Testowanie modeli**: Generowanie r√≥≈ºnorodnych profili klient√≥w do walidacji
2. **Ochrona prywatno≈õci**: Tworzenie realistycznych ale anonimowych danych
3. **Analiza segment√≥w**: Eksploracja zachowa≈Ñ r√≥≈ºnych grup klient√≥w
4. **Rozw√≥j produkt√≥w**: Symulacja reakcji na nowe produkty ubezpieczeniowe

---

## 4. PRZYK≈ÅAD 3: TRENING GAN - PROCES NAUKI
### Poziom trudno≈õci: ‚≠ê‚≠ê‚≠ê (≈öredni)

Teraz nauczymy siƒô jak trenowaƒá sieƒá GAN. To najbardziej skomplikowana czƒô≈õƒá prezentacji.
"""

def train_insurance_gan(generator, discriminator, real_data, epochs=100, batch_size=64):
    """
    Funkcja trenujƒÖca GAN dla danych ubezpieczeniowych.

    Parametry:
    - generator: sieƒá generujƒÖca dane
    - discriminator: sieƒá oceniajƒÖca dane
    - real_data: prawdziwe dane do nauki
    - epochs: liczba epok treningu
    - batch_size: rozmiar batcha
    """

    # Konwersja do tensor√≥w PyTorch
    real_data_tensor = torch.FloatTensor(real_data)

    # Optymizatory - osobne dla ka≈ºdej sieci!
    # Algorytm Adam dzia≈Ça dobrze dla GAN
    optimizer_G = optim.Adam(generator.parameters(), lr=0.0002, betas=(0.5, 0.999))
    optimizer_D = optim.Adam(discriminator.parameters(), lr=0.0002, betas=(0.5, 0.999))

    # Funkcja straty - Binary Cross Entropy
    criterion = nn.BCELoss()

    # Listy do przechowywania historii treningu
    d_losses = []
    g_losses = []

    print("üöÄ Rozpoczynanie treningu GAN...")
    print("üìä Format: [Epoka] D_loss: X.XXX, G_loss: X.XXX")

    for epoch in range(epochs):
        # Liczniki strat w epoce
        d_loss_epoch = 0
        g_loss_epoch = 0
        num_batches = 0

        # Trening w batchach
        for i in range(0, len(real_data_tensor), batch_size):
            # Pobieranie batcha prawdziwych danych
            real_batch = real_data_tensor[i:i+batch_size]
            batch_size_actual = real_batch.size(0)

            # === TRENING DYSKRYMINATORA ===
            optimizer_D.zero_grad()

            # Etykiety: 1 = prawdziwe, 0 = fa≈Çszywe
            real_labels = torch.ones(batch_size_actual, 1)
            fake_labels = torch.zeros(batch_size_actual, 1)

            # Ocena prawdziwych danych
            real_predictions = discriminator(real_batch)
            real_loss = criterion(real_predictions, real_labels)

            # Generowanie fa≈Çszywych danych
            noise = torch.randn(batch_size_actual, 100)
            fake_data = generator(noise)

            # Ocena fa≈Çszywych danych (detach() ≈ºeby nie trenowaƒá generatora!)
            fake_predictions = discriminator(fake_data.detach())
            fake_loss = criterion(fake_predictions, fake_labels)

            # ≈ÅƒÖczna strata discriminatora
            d_loss = real_loss + fake_loss
            d_loss.backward()
            optimizer_D.step()

            # === TRENING GENERATORA ===
            optimizer_G.zero_grad()

            # Generator chce, ≈ºeby dyskryminator uzna≈Ç jego dane za prawdziwe
            # Dlatego u≈ºywamy real_labels (1) jako target!
            fake_predictions = discriminator(fake_data)
            g_loss = criterion(fake_predictions, real_labels)

            g_loss.backward()
            optimizer_G.step()

            # Akumulacja strat
            d_loss_epoch += d_loss.item()
            g_loss_epoch += g_loss.item()
            num_batches += 1

        # ≈örednie straty w epoce
        d_loss_avg = d_loss_epoch / num_batches
        g_loss_avg = g_loss_epoch / num_batches

        d_losses.append(d_loss_avg)
        g_losses.append(g_loss_avg)

        # Wy≈õwietlanie postƒôpu co 10 epok
        if epoch % 10 == 0:
            print(f"[{epoch:3d}] D_loss: {d_loss_avg:.3f}, G_loss: {g_loss_avg:.3f}")

    print("‚úÖ Trening zako≈Ñczony!")
    return d_losses, g_losses

# URUCHOMIENIE TRENINGU

print("‚ö†Ô∏è  UWAGA: Trening mo≈ºe potrwaƒá 2-5 minut...")
d_losses, g_losses = train_insurance_gan(
    gen_insurance,
    disc_insurance,
    real_data_normalized,
    epochs=50,  # Mniej epok dla demonstracji
    batch_size=32
)

# Wizualizacja przebiegu treningu
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(d_losses, label='Warto≈õƒá funkcji straty dla dyskryminatora', color='red')
plt.plot(g_losses, label='Warto≈õƒá funkcji straty dla generatora', color='blue')
plt.title('Przebieg treningu GAN')
plt.xlabel('Epoka')
plt.ylabel('Strata')
plt.legend()
plt.grid(True, alpha=0.3)

# Generowanie i wizualizacja wynik√≥w
with torch.no_grad():
    test_noise = torch.randn(1000, 100)
    generated_data = gen_insurance(test_noise).numpy()

# Denormalizacja wygenerowanych danych
generated_data_denorm = scaler.inverse_transform(generated_data)

plt.subplot(1, 2, 2)
plt.scatter(real_data[:, 0], real_data[:, 1], alpha=0.5, label='Prawdziwe dane', s=10)
plt.scatter(generated_data_denorm[:, 0], generated_data_denorm[:, 1],
           alpha=0.5, label='Wygenerowane dane', s=10)
plt.xlabel('Wiek')
plt.ylabel('Doch√≥d')
plt.title('Por√≥wnanie danych: Prawdziwe a wygenerowane')
plt.legend()

plt.tight_layout()
plt.show()

print("\nüìä Statystyki por√≥wnawcze:")
print("Prawdziwe dane:")
print(pd.DataFrame(real_data, columns=feature_names).describe())
print("\nWygenerowane dane:")
print(pd.DataFrame(generated_data_denorm, columns=feature_names).describe())

"""
### Istotne obserwacje wynikajƒÖce z treningu:

1. **Balans strat**: Generator i dyskryminator powinny "walczyƒá" r√≥wnomiernie
2. **Stabilno≈õƒá**: Du≈ºe wahania mogƒÖ oznaczaƒá problemy z treningiem
3. **Jako≈õƒá danych**: Wygenerowane dane powinny przypominaƒá prawdziwe

### Praktyczne zastosowania:

- **Stress testing**: Generowanie ekstremalnych profili klient√≥w
- **Regulacje GDPR**: Zastƒôpowanie prawdziwych danych syntetycznymi
- **A/B testing**: Tworzenie kontrolowanych grup testowych

---

## 5. PRZYK≈ÅAD 4: GAN DO WYKRYWANIA OSZUSTW UBEZPIECZENIOWYCH
### Poziom trudno≈õci: ‚≠ê‚≠ê‚≠ê‚≠ê (Trudny)

Przyk≈Çad - u≈ºywamy GAN do generowania przyk≈Çad√≥w oszustw ubezpieczeniowych!
"""

def create_fraud_dataset(n_samples=1000, fraud_ratio=0.1):
    """
    Tworzy zbi√≥r danych z prawdziwymi i fa≈Çszywymi roszczeniami ubezpieczeniowymi.

    Cechy roszczenia:
    - Warto≈õƒá szkody
    - Wiek pojazdu
    - Do≈õwiadczenie kierowcy
    - Lokalizacja (kod pocztowy)
    - Czas od polisy do szkody
    - Dzie≈Ñ tygodnia zdarzenia
    """
    np.random.seed(42)

    n_fraud = int(n_samples * fraud_ratio)
    n_legitimate = n_samples - n_fraud

    # === PRAWDZIWE ROSZCZENIA ===
    # Normalne szkody - rozk≈Çad logarytmiczno-normalny
    legitimate_damage = np.random.lognormal(8, 1, n_legitimate)  # ~3k ≈õrednia
    legitimate_damage = np.clip(legitimate_damage, 500, 50000)

    # Wiek pojazdu - normalny rozk≈Çad
    legitimate_age = np.random.normal(7, 4, n_legitimate)
    legitimate_age = np.clip(legitimate_age, 1, 20)

    # Do≈õwiadczenie kierowcy
    legitimate_exp = np.random.normal(15, 8, n_legitimate)
    legitimate_exp = np.clip(legitimate_exp, 1, 40)

    # Kod pocztowy (symulowany)
    legitimate_postal = np.random.randint(10000, 99999, n_legitimate)

    # Czas od polisy - wiƒôkszo≈õƒá w pierwszych latach
    legitimate_time = np.random.exponential(1.5, n_legitimate)
    legitimate_time = np.clip(legitimate_time, 0.1, 5)

    # Dzie≈Ñ tygodnia (1-7)
    legitimate_weekday = np.random.randint(1, 8, n_legitimate)

    # === OSZUKA≈ÉCZE ROSZCZENIA ===
    # Wy≈ºsze szkody, podejrzane wzorce
    fraud_damage = np.random.lognormal(9.5, 0.8, n_fraud)  # ~13k ≈õrednia!
    fraud_damage = np.clip(fraud_damage, 5000, 100000)

    # Starsze pojazdy (≈Çatwiej oszukaƒá)
    fraud_age = np.random.normal(12, 3, n_fraud)
    fraud_age = np.clip(fraud_age, 5, 20)

    # Mniej do≈õwiadczeni kierowcy
    fraud_exp = np.random.normal(8, 5, n_fraud)
    fraud_exp = np.clip(fraud_exp, 1, 20)

    # Koncentracja w okre≈õlonych kodach (oszu≈õci siƒô grupujƒÖ)
    fraud_postal_base = np.random.choice([12345, 54321, 11111], n_fraud)
    fraud_postal = fraud_postal_base + np.random.randint(-100, 100, n_fraud)

    # Kr√≥tki czas od polisy (szybkie oszustwa)
    fraud_time = np.random.exponential(0.3, n_fraud)
    fraud_time = np.clip(fraud_time, 0.01, 1)

    # Unikanie weekend√≥w (mniej ≈õwiadk√≥w)
    fraud_weekday = np.random.choice([1,2,3,4,5], n_fraud, p=[0.3,0.3,0.2,0.1,0.1])

    # ≈ÅƒÖczenie danych
    all_damage = np.concatenate([legitimate_damage, fraud_damage])
    all_age = np.concatenate([legitimate_age, fraud_age])
    all_exp = np.concatenate([legitimate_exp, fraud_exp])
    all_postal = np.concatenate([legitimate_postal, fraud_postal])
    all_time = np.concatenate([legitimate_time, fraud_time])
    all_weekday = np.concatenate([legitimate_weekday, fraud_weekday])

    # Etykiety: 0 = legalne, 1 = oszustwo
    labels = np.concatenate([np.zeros(n_legitimate), np.ones(n_fraud)])

    # ≈ÅƒÖczenie w jeden dataset
    data = np.column_stack([all_damage, all_age, all_exp, all_postal, all_time, all_weekday])

    # Mieszanie danych
    indices = np.random.permutation(len(data))
    data = data[indices]
    labels = labels[indices]

    return data, labels

# Tworzenie zbioru danych oszustw
fraud_data, fraud_labels = create_fraud_dataset(2000, fraud_ratio=0.15)
fraud_feature_names = ['Szkoda_PLN', 'Wiek_pojazdu', 'Do≈õwiadczenie_kierowcy',
                      'Kod_pocztowy', 'Czas_od_polisy', 'Dzie≈Ñ_tygodnia']

print("Zbi√≥r danych oszustw ubezpieczeniowych utworzony!")
print(f"Rozk≈Çad: {(fraud_labels==0).sum()} legalnych, {(fraud_labels==1).sum()} oszustw")

# Analiza r√≥≈ºnic miƒôdzy grupami
fraud_df = pd.DataFrame(fraud_data, columns=fraud_feature_names)
fraud_df['Oszustwo'] = fraud_labels

print("\nPor√≥wnanie ≈õrednich warto≈õci:")
comparison = fraud_df.groupby('Oszustwo').mean()
comparison.index = ['Legalne', 'Oszustwo']
print(comparison)

# Wizualizacja r√≥≈ºnic
fig, axes = plt.subplots(2, 3, figsize=(15, 10))
fig.suptitle('R√≥≈ºnice miƒôdzy legalnymi roszczeniami a oszustwami', fontsize=16)

for i, (feature, ax) in enumerate(zip(fraud_feature_names, axes.flat)):
    legitimate = fraud_data[fraud_labels == 0, i]
    fraudulent = fraud_data[fraud_labels == 1, i]

    ax.hist(legitimate, bins=30, alpha=0.7, label='Legalne', color='green')
    ax.hist(fraudulent, bins=30, alpha=0.7, label='Oszustwo', color='red')
    ax.set_title(f'{feature}')
    ax.legend()
    ax.set_xlabel(feature)
    ax.set_ylabel('Czƒôsto≈õƒá')

plt.tight_layout()
plt.show()

class FraudGAN_Generator(nn.Module):
    """
    Generator do tworzenia syntetycznych przyk≈Çad√≥w oszustw.

    Pomaga w treningu system√≥w wykrywajƒÖcych oszustwa poprzez
    dostarczenie wiƒôkszej liczby przyk≈Çad√≥w fraud√≥w.
    """
    def __init__(self, noise_dim=100, output_dim=6):
        super(FraudGAN_Generator, self).__init__()

        # Bardziej z≈Ço≈ºona architektura dla trudniejszych danych
        self.network = nn.Sequential(
            nn.Linear(noise_dim, 256),
            nn.BatchNorm1d(256),
            nn.LeakyReLU(0.2),
            nn.Dropout(0.2),

            nn.Linear(256, 512),
            nn.BatchNorm1d(512),
            nn.LeakyReLU(0.2),
            nn.Dropout(0.2),

            nn.Linear(512, 256),
            nn.BatchNorm1d(256),
            nn.LeakyReLU(0.2),

            nn.Linear(256, output_dim),
            nn.Tanh()
        )

    def forward(self, noise):
        return self.network(noise)

class FraudGAN_Discriminator(nn.Module):
    """
    Dyskryminator wyspecjalizowany w rozpoznawaniu wzorc√≥w oszustw.
    """
    def __init__(self, input_dim=6):
        super(FraudGAN_Discriminator, self).__init__()

        self.network = nn.Sequential(
            nn.Linear(input_dim, 512),
            nn.LeakyReLU(0.2),
            nn.Dropout(0.4),

            nn.Linear(512, 256),
            nn.LeakyReLU(0.2),
            nn.Dropout(0.4),

            nn.Linear(256, 128),
            nn.LeakyReLU(0.2),
            nn.Dropout(0.3),

            nn.Linear(128, 1),
            nn.Sigmoid()
        )

    def forward(self, data):
        return self.network(data)

# Przygotowanie danych - tylko oszustwa dla generatora
fraud_only_data = fraud_data[fraud_labels == 1]
fraud_scaler = StandardScaler()
fraud_normalized = fraud_scaler.fit_transform(fraud_only_data)

print(f"\nüîç Dane do treningu GAN oszustw:")
print(f"Liczba przyk≈Çad√≥w oszustw: {len(fraud_normalized)}")
print(f"Wymiary danych: {fraud_normalized.shape}")

# Tworzenie sieci
fraud_generator = FraudGAN_Generator(noise_dim=100, output_dim=6)
fraud_discriminator = FraudGAN_Discriminator(input_dim=6)

print(f"\nSieci GAN do oszustw:")
print(f"Generator: {sum(p.numel() for p in fraud_generator.parameters())} parametr√≥w")
print(f"Discriminator: {sum(p.numel() for p in fraud_discriminator.parameters())} parametr√≥w")

# Kr√≥tki trening (demonstracyjny)
print("\nUruchamianie treningu GAN dla oszustw...")
fraud_d_losses, fraud_g_losses = train_insurance_gan(
    fraud_generator,
    fraud_discriminator,
    fraud_normalized,
    epochs=30,
    batch_size=16
)

# Generowanie syntetycznych oszustw
print("\nGenerowanie syntetycznych przyk≈Çad√≥w oszustw...")
with torch.no_grad():
    synthetic_noise = torch.randn(200, 100)
    synthetic_frauds = fraud_generator(synthetic_noise).numpy()

# Denormalizacja
synthetic_frauds_denorm = fraud_scaler.inverse_transform(synthetic_frauds)

# Por√≥wnanie prawdziwych i syntetycznych oszustw
print("\nPor√≥wnanie prawdziwych vs syntetycznych oszustw:")

real_frauds_df = pd.DataFrame(fraud_only_data, columns=fraud_feature_names)
synthetic_frauds_df = pd.DataFrame(synthetic_frauds_denorm, columns=fraud_feature_names)

print("Prawdziwe oszustwa - statystyki:")
print(real_frauds_df.describe().round(2))
print("\nSyntetyczne oszustwa - statystyki:")
print(synthetic_frauds_df.describe().round(2))

# Wizualizacja por√≥wnania
fig, axes = plt.subplots(2, 3, figsize=(15, 10))
fig.suptitle('Por√≥wnanie: Prawdziwe vs Syntetyczne oszustwa', fontsize=16)

for i, (feature, ax) in enumerate(zip(fraud_feature_names, axes.flat)):
    ax.hist(fraud_only_data[:, i], bins=20, alpha=0.7, label='Prawdziwe', color='red')
    ax.hist(synthetic_frauds_denorm[:, i], bins=20, alpha=0.7, label='Syntetyczne', color='blue')
    ax.set_title(f'{feature}')
    ax.legend()
    ax.set_xlabel(feature)
    ax.set_ylabel('Czƒôsto≈õƒá')

plt.tight_layout()
plt.show()

"""
### Korzy≈õci dla wykrywania oszustw:

1. **Wiƒôcej danych treningowych**: GAN generuje dodatkowe przyk≈Çady oszustw
2. **R√≥≈ºnorodno≈õƒá wzorc√≥w**: Syntetyczne dane pokazujƒÖ nowe kombinacje cech
3. **Balansowanie dataset'u**: Wyr√≥wnanie proporcji oszustw vs legalnych roszcze≈Ñ
4. **Testing edge cases**: Generowanie ekstremalnych przypadk√≥w oszustw

### Wa≈ºne uwagi:

- Syntetyczne dane nie zastƒôpujƒÖ prawdziwych - tylko je uzupe≈ÇniajƒÖ
- Zawsze trzeba weryfikowaƒá jako≈õƒá wygenerowanych danych
- Nale≈ºy regularnie aktualizowaƒá modele wprowadzajƒÖc nowe wzorce oszustw

---

## 6. PRZYK≈ÅAD 5: ZAAWANSOWANA ANALIZA I OCENA BIZNESOWA
### Poziom trudno≈õci: ‚≠ê‚≠ê‚≠ê‚≠ê‚≠ê (Ekspert)

Ostatni przyk≈Çad pokazuje kompleksowƒÖ ewaluacjƒô GAN i obliczenie warto≈õci biznesowej.
"""

def calculate_business_value(real_data, synthetic_data, fraud_detection_improvement):
    """
    Oblicza warto≈õƒá biznesowƒÖ implementacji GAN w firmie ubezpieczeniowej.

    Parametry:
    - real_data: prawdziwe dane
    - synthetic_data: dane wygenerowane przez GAN
    - fraud_detection_improvement: poprawa wykrywania oszustw (%)
    """

    # Symulowane warto≈õci biznesowe firmy ubezpieczeniowej
    annual_claims_volume = 50000  # roczna liczba roszcze≈Ñ
    average_claim_value = 8000    # ≈õrednia warto≈õƒá roszczenia (PLN)
    fraud_rate = 0.12            # 12% roszcze≈Ñ to oszustwa
    current_detection_rate = 0.65 # obecne wykrywanie 65% oszustw

    # Obliczenia bazowe
    annual_claims_value = annual_claims_volume * average_claim_value
    annual_fraud_value = annual_claims_value * fraud_rate
    currently_detected_fraud = annual_fraud_value * current_detection_rate
    currently_undetected_fraud = annual_fraud_value - currently_detected_fraud

    # Poprawa dziƒôki GAN
    new_detection_rate = min(0.95, current_detection_rate + fraud_detection_improvement)
    newly_detected_fraud = annual_fraud_value * new_detection_rate
    additional_savings = newly_detected_fraud - currently_detected_fraud

    # Koszty implementacji
    implementation_cost = 500000   # PLN - jednorazowy koszt
    annual_maintenance = 100000    # PLN - roczne utrzymanie

    # ROI calculation
    net_annual_benefit = additional_savings - annual_maintenance
    roi_year_1 = (net_annual_benefit - implementation_cost) / implementation_cost * 100
    roi_year_3 = ((net_annual_benefit * 3) - implementation_cost) / implementation_cost * 100

    results = {
        'Roczna warto≈õƒá roszcze≈Ñ': annual_claims_value,
        'Roczna warto≈õƒá oszustw': annual_fraud_value,
        'Obecnie wykrywane oszustwa': currently_detected_fraud,
        'Nowa stopa wykrywania': f"{new_detection_rate:.1%}",
        'Dodatkowe oszczƒôdno≈õci roczne': additional_savings,
        'Koszt implementacji': implementation_cost,
        'Roczne utrzymanie': annual_maintenance,
        'ROI rok 1': f"{roi_year_1:.1f}%",
        'ROI rok 3': f"{roi_year_3:.1f}%",
        'Payback period': implementation_cost / net_annual_benefit
    }

    return results

def evaluate_gan_quality(real_data, synthetic_data, feature_names):
    """
    Ocenia jako≈õƒá danych wygenerowanych przez GAN u≈ºywajƒÖc r√≥≈ºnych metryk.
    """
    from scipy import stats

    print("üîç OCENA JAKO≈öCI DANYCH WYGENEROWANYCH PRZEZ GAN")
    print("=" * 60)

    quality_scores = {}

    # 1. Por√≥wnanie rozk≈Çad√≥w statystycznych
    print("\nüìä 1. ANALIZA ROZK≈ÅAD√ìW STATYSTYCZNYCH")
    for i, feature in enumerate(feature_names):
        real_feature = real_data[:, i]
        synthetic_feature = synthetic_data[:, i]

        # Test Ko≈Çmogorowa-Smirnowa
        ks_statistic, ks_p_value = stats.ks_2samp(real_feature, synthetic_feature)

        # Por√≥wnanie moment√≥w
        real_mean, real_std = np.mean(real_feature), np.std(real_feature)
        synth_mean, synth_std = np.mean(synthetic_feature), np.std(synthetic_feature)

        mean_diff = abs(real_mean - synth_mean) / real_mean * 100
        std_diff = abs(real_std - synth_std) / real_std * 100

        print(f"\n{feature}:")
        print(f"  KS test p-value: {ks_p_value:.4f} {'‚úÖ' if ks_p_value > 0.05 else '‚ùå'}")
        print(f"  R√≥≈ºnica ≈õrednich: {mean_diff:.1f}%")
        print(f"  R√≥≈ºnica odchyle≈Ñ: {std_diff:.1f}%")

        # Scoring (0-100)
        ks_score = 100 if ks_p_value > 0.05 else max(0, 100 * ks_p_value / 0.05)
        mean_score = max(0, 100 - mean_diff)
        std_score = max(0, 100 - std_diff)

        feature_score = (ks_score + mean_score + std_score) / 3
        quality_scores[feature] = feature_score

    # 2. Analiza korelacji
    print(f"\nüìà 2. ANALIZA KORELACJI")
    real_corr = np.corrcoef(real_data.T)
    synthetic_corr = np.corrcoef(synthetic_data.T)

    corr_diff = np.abs(real_corr - synthetic_corr)
    avg_corr_diff = np.mean(corr_diff[np.triu_indices_from(corr_diff, k=1)])

    print(f"≈örednia r√≥≈ºnica korelacji: {avg_corr_diff:.3f}")
    print(f"Ocena korelacji: {'‚úÖ' if avg_corr_diff < 0.1 else '‚ùå'}")

    # 3. Wykrywanie outlier√≥w
    print(f"\nüéØ 3. ANALIZA OUTLIER√ìW")

    def count_outliers(data):
        Q1 = np.percentile(data, 25, axis=0)
        Q3 = np.percentile(data, 75, axis=0)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        outliers = np.sum((data < lower_bound) | (data > upper_bound))
        return outliers

    real_outliers = count_outliers(real_data)
    synthetic_outliers = count_outliers(synthetic_data)

    outlier_ratio = synthetic_outliers / real_outliers if real_outliers > 0 else 1
    print(f"Outliers w prawdziwych danych: {real_outliers}")
    print(f"Outliers w syntetycznych danych: {synthetic_outliers}")
    print(f"Stosunek outlier√≥w: {outlier_ratio:.2f}")

    # 4. Ocena og√≥lna
    overall_quality = np.mean(list(quality_scores.values()))

    print(f"\n OCENA OG√ìLNA")
    print(f"Jako≈õƒá danych GAN: {overall_quality:.1f}/100")

    if overall_quality >= 80:
        grade = " Doskona≈Ça - gotowa do produkcji"
    elif overall_quality >= 70:
        grade = " Bardzo dobra - wymaga drobnych poprawek"
    elif overall_quality >= 60:
        grade = " Dobra - wymaga optymalizacji"
    else:
        grade = " NiezadowalajƒÖca - wymaga przebudowy"

    print(f"Rekomendacja: {grade}")

    return quality_scores, overall_quality

# OCENA JAKO≈öCI WSZYSTKICH NASZYCH MODELI
print("üß™ KOMPLEKSOWA OCENA JAKO≈öCI MODELI GAN")
print("=" * 80)

# Ocena modelu do danych klient√≥w
print("\n  OCENA MODELU DANYCH KLIENT√ìW")
with torch.no_grad():
    # U≈ºywamy odpowiedniego wymiaru szumu (100 dla gen_insurance)
    client_noise = torch.randn(1000, 100)
    client_synthetic = gen_insurance(client_noise).numpy()
    client_synthetic_denorm = scaler.inverse_transform(client_synthetic)

client_scores, client_quality = evaluate_gan_quality(
    real_data, client_synthetic_denorm, feature_names
)

# Ocena modelu oszustw (ju≈º mamy dane)
print("\n  OCENA MODELU OSZUSTW")
fraud_scores, fraud_quality = evaluate_gan_quality(
    fraud_only_data, synthetic_frauds_denorm, fraud_feature_names
)

# ANALIZA WARTO≈öCI BIZNESOWEJ
print("\n ANALIZA WARTO≈öCI BIZNESOWEJ")
print("=" * 80)

# Szacujemy poprawƒô wykrywania oszustw na podstawie jako≈õci modelu
fraud_improvement = (fraud_quality / 100) * 0.15  # max 15% poprawa

business_value = calculate_business_value(
    real_data=fraud_data,
    synthetic_data=synthetic_frauds_denorm,
    fraud_detection_improvement=fraud_improvement
)

print("\n KLUCZOWE WSKA≈πNIKI BIZNESOWE:")
for key, value in business_value.items():
    if isinstance(value, (int, float)) and key not in ['ROI rok 1', 'ROI rok 3', 'Nowa stopa wykrywania']:
        print(f"{key}: {value:,.0f} PLN")
    else:
        print(f"{key}: {value}")

# Wizualizacja warto≈õci biznesowej
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(15, 12))
fig.suptitle('Analiza Warto≈õci Biznesowej Implementacji GAN', fontsize=16)

# Wykres 1: Oszczƒôdno≈õci w czasie
years = [1, 2, 3, 4, 5]
annual_savings = business_value['Dodatkowe oszczƒôdno≈õci roczne']
cumulative_savings = [annual_savings * year - business_value['Koszt implementacji']
                     - business_value['Roczne utrzymanie'] * year for year in years]

ax1.bar(years, cumulative_savings, color=['red' if x < 0 else 'green' for x in cumulative_savings])
ax1.axhline(y=0, color='black', linestyle='--', alpha=0.5)
ax1.set_title('Skumulowane oszczƒôdno≈õci w czasie')
ax1.set_xlabel('Rok')
ax1.set_ylabel('Oszczƒôdno≈õci (PLN)')
ax1.grid(True, alpha=0.3)

# Wykres 2: Por√≥wnanie jako≈õci modeli
models = ['Dane klient√≥w', 'Wykrywanie oszustw']
qualities = [client_quality, fraud_quality]

bars = ax2.bar(models, qualities, color=['skyblue', 'lightcoral'])
ax2.set_title('Jako≈õƒá modeli GAN')
ax2.set_ylabel('Jako≈õƒá (%)')
ax2.set_ylim(0, 100)

# Dodanie warto≈õci na s≈Çupkach
for bar, quality in zip(bars, qualities):
    height = bar.get_height()
    ax2.text(bar.get_x() + bar.get_width()/2., height + 1,
            f'{quality:.1f}%', ha='center', va='bottom')

# Wykres 3: Rozk≈Çad koszt√≥w vs korzy≈õci (rok 1)
labels = ['Koszty implementacji', 'Roczne utrzymanie', 'Oszczƒôdno≈õci']
costs_benefits = [
    business_value['Koszt implementacji'],
    business_value['Roczne utrzymanie'],
    business_value['Dodatkowe oszczƒôdno≈õci roczne']
]
colors = ['red', 'orange', 'green']

ax3.pie(np.abs(costs_benefits), labels=labels, colors=colors, autopct='%1.1f%%')
ax3.set_title('Struktura koszt√≥w i korzy≈õci (rok 1)')

# Wykres 4: Scenariusze ROI
detection_improvements = np.linspace(0.05, 0.20, 10)
roi_scenarios = []

for improvement in detection_improvements:
    scenario = calculate_business_value(fraud_data, synthetic_frauds_denorm, improvement)
    roi_year_3 = float(scenario['ROI rok 3'].replace('%', ''))
    roi_scenarios.append(roi_year_3)

ax4.plot(detection_improvements * 100, roi_scenarios, marker='o', linewidth=2, markersize=6)
ax4.axhline(y=0, color='red', linestyle='--', alpha=0.5)
ax4.axvline(x=fraud_improvement * 100, color='green', linestyle=':', alpha=0.7,
           label=f'Obecny model: {fraud_improvement*100:.1f}%')
ax4.set_title('ROI w zale≈ºno≈õci od poprawy wykrywania oszustw')
ax4.set_xlabel('Poprawa wykrywania oszustw (%)')
ax4.set_ylabel('ROI 3 lata (%)')
ax4.grid(True, alpha=0.3)
ax4.legend()

plt.tight_layout()
plt.show()

"""
---

## 7. PODSUMOWANIE I REKOMENDACJE BIZNESOWE

###  G≈Å√ìWNE WNIOSKI

1. **Techniczne osiƒÖgniƒôcia:**
   - Uda≈Ço siƒô stworzyƒá funkcjonalne sieci GAN dla danych ubezpieczeniowych
   - Jako≈õƒá generowanych danych: {client_quality:.1f}% (klienci), {fraud_quality:.1f}% (oszustwa)
   - Modele nadajƒÖ siƒô do zastosowa≈Ñ praktycznych

2. **Warto≈õƒá biznesowa:**
   - Przewidywane roczne oszczƒôdno≈õci: {business_value['Dodatkowe oszczƒôdno≈õci roczne']:,.0f} PLN
   - ROI po 3 latach: {business_value['ROI rok 3']}
   - Okres zwrotu inwestycji: {business_value['Payback period']:.1f} lat

3. **Kluczowe zastosowania:**
   -  **Wykrywanie oszustw**: Generowanie przyk≈Çad√≥w do treningu algorytm√≥w
   -  **Ochrona prywatno≈õci**: Syntetyczne dane zamiast prawdziwych danych klient√≥w
   -  **Augmentacja danych**: Powiƒôkszanie zbior√≥w treningowych
   -  **Testowanie modeli**: Walidacja na r√≥≈ºnorodnych scenariuszach

###  REKOMENDACJE IMPLEMENTACYJNE

**FAZA 1: Pilot (3 miesiƒÖce)**
- Implementacja GAN dla wykrywania oszustw w jednym regionie
- Por√≥wnanie skuteczno≈õci z obecnymi metodami
- Szkolenie zespo≈Çu data science

**FAZA 2: Rozszerzenie (6 miesiƒôcy)**
- Wdro≈ºenie na pe≈ÇnƒÖ skalƒô dla wykrywania oszustw
- Dodanie generowania danych klient√≥w dla cel√≥w analitycznych
- Integracja z istniejƒÖcymi systemami

**FAZA 3: Optymalizacja (12 miesiƒôcy)**
- CiƒÖg≈Çe uczenie i doskonalenie modeli
- Rozszerzenie na nowe produkty ubezpieczeniowe
- Automacja proces√≥w

### ELEMENTY RYZYKA I ICH ≈ÅAGODZENIE

**Ryzyko techniczne:**
- Niestabilno≈õƒá treningu GAN ‚Üí Regularne monitorowanie i checkpointy
- Overfitting ‚Üí Walidacja na niezale≈ºnych zbiorach danych

**Ryzyko biznesowe:**
- Regulacje GDPR ‚Üí Konsultacje prawne, dokumentacja proces√≥w
- Akceptacja u≈ºytkownik√≥w ‚Üí Stopniowe wdro≈ºenie, szkolenia

**Ryzyko operacyjne:**
- Zale≈ºno≈õƒá od ekspert√≥w ‚Üí Budowanie wewnƒôtrznych kompetencji
- Koszty obliczeniowe ‚Üí Optymalizacja modeli, cloud computing

###  NASTƒòPNE KROKI

1. **Natychmiastowe (1-2 tygodnie):**
   - Prezentacja wynik√≥w zarzƒÖdowi
   - Alokacja bud≈ºetu i zasob√≥w
   - Wyb√≥r zespo≈Çu projektowego

2. **Kr√≥tkoterminowe (1-3 miesiƒÖce):**
   - Setup ≈õrodowiska produkcyjnego
   - Przygotowanie danych treningowych
   - Rozpoczƒôcie pilotu

3. **≈örednioterminowe (3-12 miesiƒôcy):**
   - Pe≈Çne wdro≈ºenie systemu
   - Szkolenie u≈ºytkownik√≥w ko≈Ñcowych
   - Monitoring i optymalizacja

###  WSPARCIE I DALSZY ROZW√ìJ

Ten notebook mo≈ºe stanowiƒá pierwszy krok dla implementacji GAN w firmie ubezpieczeniowej.
Ka≈ºdy z przedstawionych modeli mo≈ºe byƒá dalej dostosowywany do specyficznych
potrzeb biznesowych.


---
*Dokument przygotowany z wykorzystaniem najnowszych technologii sztucznej inteligencji*
*dla potrzeb strategicznych firmy ubezpieczeniowej - Wrzesie≈Ñ 2025*
"""

print(" PREZENTACJA ZAKO≈ÉCZONA!")
print("=" * 80)
print(" Wszystkie przyk≈Çady kodu zosta≈Çy pomy≈õlnie wykonane")
print(" Analiza biznesowa zosta≈Ça przeprowadzona")
print(" Rekomendacje sƒÖ gotowe do przedstawienia zarzƒÖdowi")
print("\n Dalsze eksperymenty mo≈ºna prowadziƒá modyfikujƒÖc parametry sieci,")
print("   zmieniajƒÖc architektury lub dodajƒÖc nowe ≈∫r√≥d≈Ça danych.")
print("\n Ten notebook mo≈ºe s≈Çu≈ºyƒá jako szablon dla podobnych projekt√≥w")
print("   w innych dzia≈Çach firmy ubezpieczeniowej.")
```
